```python
import os
import glob
import pandas as pd
```

```python
folder_path = 'C:/Users/johfr/Dropbox/Skule/Data science/Bachelor/Dataset/Mirai/'
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))  
```

```python
df_list = []
for file in csv_files:
    df = pd.read_csv(file).sample(n=1000, random_state=42)
    df_list.append(df)
```

```python
combined_df = pd.concat(df_list, ignore_index=True)
output_file = os.path.join(folder_path, 'combined_mirai.csv')
combined_df.to_csv(output_file, index=False)
```
